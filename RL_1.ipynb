Co-relation
Positive- if one increase all value will increase
Negative - if one decreases all vakue will decrease
Zero- there is no co-relation

```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0,50,1000)
```

```python
#Positive correlation with some noise
y = x + np.random.normal(0, 10, 1000)
print(y)
```

```python
#Positive correlation with some noise
y = x + np.random.normal(0, 10, 1000)
print(y)
```

```python
np.corrcoef(x,y)
```

this correlation is 0.815 a strong positive correlation

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

x axis side graph is positive relation

```python
import pandas as pd
df = pd.DataFrame({'a' : np.random.randint(0, 50, 1000)})
df['b'] = df['a'] + np.random.normal(0, 10, 1000) #positively correlated with 'a'
df['c'] = 100 - df['a'] + np.random.normal(0, 5, 1000) #negatively correlated with 'a'
df['d'] = np.random.randint(0, 50, 1000) #not correlated with 'a'

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize=(6, 6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
